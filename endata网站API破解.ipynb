```python
#@Author:Steve Heydrich, South China University of Technology
#@CreateTime: 2024/12/02
#@NB:本代码仅供华南理工大学电子商务系数据采集及网络爬虫课程学习使用，以此盈利的雨我无瓜
import subprocess
import requests
import json 
import pandas as pd
url="https://www.endata.com.cn/API/GetData.ashx" #API的URL
#向API URL发送请求，data参数就是post所要附带的参数
r=requests.post(url,data={"year":"2023","MethodName":"BoxOffice_GetYearInfoData"})
if r.ok:
    #API URL返回的是加密后的文本，先存到user_input变量里
    user_input = r.text
    #编写好向解密js(e.js)通讯的命令
    command = ['node', 'e.js', user_input]
    #一些设置
    process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, encoding='utf-8')
    #取回结果
    stdout, stderr = process.communicate()
    if process.returncode == 0:
        #若成功则将结果保存至变量，返回的结果是json字符串，所以用json.loads解析
        print("结果已保存至result变量")
        result=json.loads(stdout)
        print(pd.DataFrame(result['Data']['Table']))
    else:
        #报错提示
        print("JavaScript 错误:", stderr)
```
